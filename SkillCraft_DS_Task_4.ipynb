# Check Time of Day Patterns

```python
import matplotlib.pyplot as plt

df['Time of Day'].value_counts().plot(kind='bar', figsize=(8,5), color="skyblue")
plt.title("Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

# Accidents by Day of Week

```python
df['Day of Week'].value_counts().plot(kind='bar', figsize=(8,5), color="orange")
plt.title("Accidents by Day of Week")
plt.xlabel("Day")
plt.ylabel("Number of Accidents")
plt.show()
```

# Accidents by Month

```python
df['Month'].value_counts().sort_index().plot(kind='line', marker='o', figsize=(8,5))
plt.title("Accidents by Month")
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.show()
```

# Weather Conditions Impact

```python
df['Weather Conditions'].value_counts().head(10).plot(kind='barh', figsize=(8,5), color="green")
plt.title("Top 10 Weather Conditions in Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()
```

# Road Condition Impact

```python
df['Road Condition'].value_counts().plot(kind='barh', figsize=(8,5), color="red")
plt.title("Accidents by Road Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Condition")
plt.show()
```

# Accident Severity Distribution

```python
df['Accident Severity'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Accident Severity Distribution")
plt.ylabel("")
plt.show()
```
